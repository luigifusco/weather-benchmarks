```python
import xarray as xr
import numcodecs
from numcodecs.abc import Codec
import numpy as np
```

```python

class Zeros(Codec):
    codec_id = 'zeros'
    
    def __init__(self):
        pass

    def encode(self, buf):
        return buf

    def decode(self, buf, out=None):
        assert out is not None
        
        out[:] = 0

    def get_config(self):
        return {'id': self.codec_id}

    @classmethod
    def from_config(cls, config):
        return cls()

numcodecs.register_codec(Zeros)
```
